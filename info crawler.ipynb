```python
def getdt(date_v):
    try:
        
        resp = requests.get(url.format(date_v))
        resp.encoding = 'utf-8'

        add_date_dt = np.insert(np.array(ast.literal_eval(
            (BeautifulSoup(resp.text, 'lxml').find('p')).text)['data'][0:]),0,(date_v),1) #add date
        
        df1 = pd.DataFrame(np.delete(add_date_dt, np.s_[6,7,8,12], axis=1),
                                           columns = ['Date','code', "company",
                                            'fini_long', 'fini_short',"net_fini",
                                            'itc_long', 'itc_short', "net_itc",
                                            'prop_long', 'prop_short', "net_prop",
                                            'warr_long', 'warr_short',"net_warr",
                                            'net_total'])

        locale.setlocale(locale.LC_NUMERIC, '')
        df2 = pd.concat([df1[['Date','code', "company"]],
                         df1[[ 'fini_long', 'fini_short',"net_fini",
                                            'itc_long', 'itc_short', "net_itc",
                                            'prop_long', 'prop_short', "net_prop",
                                            'warr_long', 'warr_short',"net_warr",
                                            'net_total']].applymap(atoi)],axis=1)
        
        # download & clean daily dataframe, next step should be add df into db
        
        df2.to_csv('20181014_2.csv', mode='a', header=False)

        print(df2.iloc[0]['Date'])
        
        del df1[:]
        del df2[:]
        
        
    except:
        return None  # try & except: avoid holiday
```

```python
today =date.today()
for i  in range(1,150):
    today = today + timedelta(days=-1)
    date_v = (str(today).replace('-',''))
    
    time.sleep(random.uniform(0.9998,1.1007 ))
    if (i%3) == 0:
        time.sleep(random.uniform(1.3, 1.41))
    elif (i%7) == 0:
        time.sleep(random.uniform(1.35, 1.7))   
    elif (i%29) == 0:
        time.sleep(random.uniform(2.35, 3.7)) 
    
    getdt(date_v)
```
